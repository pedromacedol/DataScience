```python
DryBean = open("References/Dry_Bean_Dataset.csv", "r") #r = read
```

```python
DryBean_data = DryBean.read()
```

```python
rows_data = DryBean_data.split('\n')
```

```python
rows_data
```

```python
full_data = []
```

```python
for row in rows_data:
    row_now = row.split(',')
    full_data.append(row_now)
```
